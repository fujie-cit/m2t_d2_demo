```python
import util
import matplotlib.pyplot as plt
```

```python
d2predictor = util.Detectron2Predictor()
m2transformer = util.M2TransformerD2()
visualizer = util.D2ResultVisualizer()
```

```python
image = util.imread_web("https://pakutaso.cdn.rabify.me/shared/img/thumb/bobjinno705P6200323.jpg.webp?d=1420")
```

```python
instances, features = d2predictor.predict(image)
sentences = m2transformer.predict(features, out_size=30)

visualizer.visualize(image, instances)
plt.show()
print("\n".join([''.join(s.split(' ')) for s in sentences]))
```
